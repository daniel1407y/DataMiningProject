Since we assumed that zip_code is correct (and corrected it), and fixed state_code and city respectively, we now will change longitude and latitude respectively to the zip_code

```python
search = SearchEngine()

for i, row in df.iterrows():
    state = row['state_code']
    city = row['city']
    zip_code = row['zip_code']
    latitude = row['latitude']
    longitude = row['longitude']
    
    result = search.by_zipcode(zip_code)
    
    if result:
        if abs(result.lat - latitude) > 0.1 or abs(result.lng - longitude) > 1:
            print(f"Mismatch at index {i}: Coordinates for {city}, {state} ({latitude}, {longitude}) do not match with zip code ({result.lat}, {result.lng}), Changing longitude and latitude of the startup")
            df.at[i, 'latitude'] = result.lat
            df.at[i, 'longitude'] = result.lng
    else:
        print(f"No result found for zip code {zip_code}")
```

We noticed that alot of startups have their first/last foundation dates being before their foundation date, meaning that they recieved funding before even being founded. after a search online (and a bit of common sense), we concluded that such a thing cant happen, because funding is effectively buying a share of the company.

We will work on foundation_date and first_founding_date  & foundation_date and last_founding_date

*   If the year and month are the same, change last_founding_date day to be the same as foundation_date. 
*   If the year and day are the same, change last_founding_date month to be the same as foundation_date.
*   If the month and day are the same, change last_founding_date year to be the same as foundation_date.


```python
df[['foundation_date', 'first_funding_date']] = df[['foundation_date' , 'first_funding_date']].apply(pd.to_datetime, format='%d/%m/%Y')

for i, row in df.loc[df['negative_founding_and_first_date'] == 1].iterrows():
    if row['foundation_date'].year == row['first_funding_date'].year and row['foundation_date'].month == row['first_funding_date'].month:
        day = row['foundation_date'].day
        df.at[i, 'first_funding_date'] = row['first_funding_date'].replace(day=day)
        df.at[i, 'negative_founding_and_first_date'] = 0
    elif row['foundation_date'].year == row['first_funding_date'].year and row['foundation_date'].day == row['first_funding_date'].day:
        month = row['foundation_date'].month
        df.at[i, 'first_funding_date'] = row['first_funding_date'].replace(month=month)
        df.at[i, 'negative_founding_and_first_date'] = 0
    elif row['foundation_date'].month == row['first_funding_date'].month and row['foundation_date'].day == row['first_funding_date'].day:
        year = row['foundation_date'].year
        df.at[i, 'first_funding_date'] = row['first_funding_date'].replace(year=year)
        df.at[i, 'negative_founding_and_first_date'] = 0 
```

```python
print("Startups with first_funding_date < foundation_date:\n", df.loc[df['negative_founding_and_first_date']==1, ['foundation_date', 'first_funding_date']])
print("total: ", len(df[df['first_funding_date'] < df['foundation_date']]))
```

## Categories

We can see that our categories have a lot of columns, represented in a boolean 
style. We will create a new column called 'main_categories', and write the correct category there:

```python
# creating a new column for categories
def calculate_new_main_category(row):
    if row['is_software']==1:
      return "software"
    elif row['is_web']==1:
      return "web" 
    elif row['is_mobile']==1:
      return "mobile"
    elif row['is_enterprise']==1:
      return "enterprise"
    elif row['is_advertising']==1:
      return "advertising"
    elif row['is_gamesvideo']==1:
      return "gamesvideo"
    elif row['is_ecommerce']==1:
      return "ecommerce"
    elif row['is_biotech']==1:
      return "biotech"
    elif row['is_consulting']==1:
      return "consulting"
    elif row['is_othercategory']==1:
      return "other"
    else:
      return "error"
    
df['main_categories'] = df.apply(calculate_new_main_category, axis=1)
```
